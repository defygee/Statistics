```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
#2.1 Построим функцию экспоненциального распределения
expon_rv = sts.expon()
expon_rv1 = sts.expon(0.6)
expon_rv2 = sts.expon(2,1)


expon_rv.rvs(size=10)
x = np.linspace(0,8)
cdf = expon_rv.cdf(x) 
cdf1 = expon_rv1.cdf(x)
cdf2 = expon_rv2.cdf(x)
plt.plot(x, cdf)
plt.plot(x, cdf1)
plt.plot(x, cdf2)

plt.legend(['0:0',  '0.6:0', '2:1'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
#2.2 Построим плотность экспоненциального распределения
expon_rv = sts.expon()
expon_rv3 = sts.expon(0, 2)
expon_rv4 = sts.expon(3, 1)
x = np.linspace(0,10,100)
pdf = expon_rv.pdf(x)
pdf3 = expon_rv3.pdf(x)
pdf4 = expon_rv4.pdf(x)
plt.figure(figsize=(15,8))
plt.plot(x, pdf)

plt.plot(x, pdf3)
plt.plot(x, pdf4)

plt.legend(['1:7',  '0:2', '3:1'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#1.2 Построим функцию распределения Пуассона
poisson_rv = sts.poisson(10)
poisson_rv1 = sts.poisson(5,3)
poisson_rv2 = sts.poisson(4)
```

```python
x = np.linspace(0,20,21)

cdf = poisson_rv.cdf(x)
cdf1 = poisson_rv1.cdf(x)
cdf2 = poisson_rv2.cdf(x)

plt.plot(x, cdf)
plt.plot(x, cdf1)
plt.plot(x, cdf2)

plt.legend(['10:0',  '5:3', '4:0'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
#1.2 Построим плотность распределения Пуассона
x = np.linspace(0,20,21)
pmf = poisson_rv.pmf(x)
pmf1 = poisson_rv1.pmf(x)
pmf2 = poisson_rv2.pmf(x)

plt.plot(x, pmf, 'o')
plt.plot(x, pmf1, 'o')
plt.plot(x, pmf2, 'o')

plt.legend(['10:0',  '5:3', '4:0'])
plt.ylabel('$P(X=x)$')
plt.xlabel('$x$')
```

```python
#3.1 Построим функцию распределения хи^2
chi2_rv = sts.chi2(2,3)
chi2_rv1 = sts.chi2(0.6)
chi2_rv2 = sts.chi2(2,1)


chi2_rv.rvs(size=10)
x = np.linspace(0,8)
cdf = chi2_rv.cdf(x) 
cdf1 = chi2_rv1.cdf(x)
cdf2 = chi2_rv2.cdf(x)
plt.plot(x, cdf)
plt.plot(x, cdf1)
plt.plot(x, cdf2)

plt.legend(['2:3',  '0.6:0', '2:1'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
#3.1 Построим плотность распределения хи^2
chi2_rv = sts.chi2(2,3)
chi2_rv3 = sts.chi2(0.6)
chi2_rv4 = sts.chi2(2, 1)
x = np.linspace(0,10,100)
pdf = chi2_rv.pdf(x)
pdf3 = chi2_rv3.pdf(x)
pdf4 = chi2_rv4.pdf(x)
plt.figure(figsize=(15,8))
plt.plot(x, pdf)

plt.plot(x, pdf3)
plt.plot(x, pdf4)

plt.legend(['2:3',  '0.6:0', '2:1'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
